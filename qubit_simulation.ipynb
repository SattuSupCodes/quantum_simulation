```python
import pennylane as qml
```

```python
dev = qml.device(name="default.qubit", wires = 1)
```

```python
@qml.qnode(device=dev)
def qubit():
  qml.X(wires=0) #flipping (NOT gate)
  state = qml.state()
  return state
```

```python
print(qubit())
```
